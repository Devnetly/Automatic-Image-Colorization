```python
from PIL import Image
import os
```

```python
def is_grey_scale(img_path):
    img = Image.open(img_path).convert('RGB')
    w, h = img.size
    for i in range(w):
        for j in range(h):
            r, g, b = img.getpixel((i,j))
            if r != g != b: 
                return False
    return True
```

```python
folder = "D:\\AIDS\\S2\\Project\\coco-2017-dataset\\coco2017\\train2017"
gs_images = []
for filename in os.listdir(folder):
    if filename.endswith(".jpg"):
        image_path = os.path.join(folder, filename)
        if is_grey_scale(image_path):
            print(f"{filename} is a grayscale image")
            gs_images.append(filename)
```

```python
print(f"Total grayscale images: {len(gs_images)}")
```

```python
#delete grayscale images
for filename in gs_images:
    os.remove(os.path.join(folder, filename))
    print(f"Deleted {filename}")
```
